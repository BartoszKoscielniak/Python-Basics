# Demonstracja biblioteki (pakietu) SQLAlchemy

### SQLAlchemy to biblioteka programistyczna napisana w języku Python i służąca do realizacji dostępu do relacyjnych baz danych oraz mapowania obiektowo-relacyjnego. 

Wspiera takie bazy danych jak: Firebird, Microsoft SQL Server, MySQL, Oracle, PostgreSQL, SQLite oraz Sybase. Głównym jej celem jest wykonywanie zapytań z relacyjnej bazy danych. Jednak wyniki tych zapytań mogą być uzyskane w postaci zbiorów danych biblioteki Pandas (obiekty DataFrame) co powoduje, że z użyciem SQLAlchemy pracujemy tak, jakbyśmy wczytywali dane z plików, choć są pobierane z relacyjnej bazy danych.

Instalacja pakietu SQLAlchemy odbywa sie poleceniem: `pip install SQLAlchemy`

Aby dostawać się do bazy PostgreSQL trzeba zainstalować jeszcze driver: `pip install psycopg2`


```python
#PIERWSZA metoda dostępu do tabeli bazy danych
#Pobieranie z bazy wszystkich wierszy jednocześnie
#Po wykonaniu zapytania, wszystkie wiersze razem są w pamięci na liście wierszy

#Polecenie usuwa wszystkie zmienne z pamięci 
%reset -f 

import sqlalchemy as db

#Dane dostępowe do bazy danych
dialect = "postgresql"
driver = "psycopg2"

database = "JUPTEST"

CzyLOKALNIE = False

if CzyLOKALNIE: #Lokalnie
    username = "postgres" 
    password = "polska2019"
    host = "localhost"
    port = "5432"
else: #Odlegly    
    username = "student" 
    password = "s#x$a2!s3y4"    
    host = "localhost"    
    port = "5433" #Inny port

    
    
engine = db.create_engine(dialect+"+"+driver+"://"+username+":"+password+"@"+host+":"+port+"/"+database)

query = "select * from SERCE limit 3"

rp = engine.execute(query)

columns = rp.keys() #Pobranie informacji o kolumnach 
print("Kolumny:",columns)

rows = rp.fetchall() #Pobranie wszystkich wierszy od razu

#Przegladanie wierszy
for row in rows:    
    i = 0
    text = ''
    #Przeglądanie kolumn w wierszu
    while i < len(row):
        value = str(row[i])
        text = text + str(value)
        if i<len(row)-1: text = text + "|"
        i = i + 1
    print(text)
```
